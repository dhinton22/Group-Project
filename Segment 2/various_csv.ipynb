```python
import pandas as pd
import numpy as np
```

```python
file_path = "../Resources/dc_crime_data.csv"
df_crime = pd.read_csv(file_path, index_col=0)
df_crime.head(5)
```

```python
# Find null values in each column
[[column,df_crime[column].isnull().sum()] for column in df_crime.columns]
```

```python
df_crime = df_crime.dropna()
df_crime. info()
```

```python
[[column,df_crime[column].isnull().sum()] for column in df_crime.columns]
```

```python
crime_areas_df = df_crime[['OFFENSE', 'CRIME_TYPE', 'CCN', 'WARD', 'LATITUDE', 'LONGITUDE']].copy()
crime_areas_df.head()
```

```python
crime_rate_year_df = df_crime[['OFFENSE', 'YEAR', 'MONTH', 'DAY']].copy()
crime_rate_year_df = crime_rate_year_df.reset_index(drop=True)
crime_rate_year_df.head()
```

```python
crime_per_psa_df = df_crime[['PSA', 'OFFENSE', 'SHIFT']].copy()
crime_per_psa_df = crime_per_psa_df.reset_index(drop=True)
crime_per_psa_df.head()
```

```python
crime_areas_df = crime_areas_df.reset_index(drop=True)
crime_areas_df.head()
```
